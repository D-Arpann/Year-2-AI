```python
# Task 24: Prove A * A^-1 = I
A = np.array([[3,4],[7,8]])
A @ np.linalg.inv(A)
```

```python
# Task 25: Show AB != BA
B = np.array([[5,3],[2,1]])
(A @ B, B @ A)
```

```python
# Task 26: (AB)^T = B^T A^T
((A @ B).T, B.T @ A.T)
```

```python
# Task 27: Solve system using inverse
A = np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
Bv = np.array([-1,-3,9])
np.linalg.inv(A).dot(Bv)
```

```python
# Task 28: Solve using linalg.solve
np.linalg.solve(A, Bv)
```
