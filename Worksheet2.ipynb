## **Medical_student.csv**

```python
# Task 1: Load medical_student.csv

print("Upload medical_student.csv")
uploaded = files.upload()

medical_df = pd.read_csv("medical_students_dataset.csv")

medical_df.head()
```

```python
# Task 2: Check for missing values

# Showing column names, dtypes, and non-null counts
medical_df.info()

print("\nMissing value count per column:")
print(medical_df.isnull().sum())
```

```python
# Task 3: Fill missing values

medical_df["Student ID"] = medical_df.index + 1

numeric_mean_cols = [
    "Age", "Height", "Weight", "BMI",
    "Temperature", "Heart Rate",
    "Blood Pressure"
]

for col in numeric_mean_cols:
    medical_df[col] = medical_df[col].fillna(medical_df[col].mean())

medical_df["Cholesterol"] = medical_df["Cholesterol"].fillna(medical_df["Cholesterol"].median())

categorical_cols = ["Gender", "Blood Type", "Diabetes", "Smoking"]

for col in categorical_cols:
    medical_df[col] = medical_df[col].fillna(medical_df[col].mode()[0])

print("\nMissing values AFTER imputation:\n")
print(medical_df.isnull().sum())
```

### Explanation

#### 1. Student ID (ffill)
Student IDs are identifiers and must follow a consistent increasing pattern. So index +1 is used.

#### 2. Numerical Columns (Mean)
Continuous columns such as Age, Height, Weight, BMI, Temperature, Heart Rate, and Blood Pressure follow normal distributions.  
Filling missing values with the **mean** preserves the central tendency of the data.

#### 3. Cholesterol (Median)
Cholesterol measurements are usually **skewed** (many high-risk patients).  
To avoid distortion caused by outliers, we use the **median**.

#### 4. Categorical Columns (Mode)
Gender, Blood Type, Diabetes status, and Smoking status are categorical.  
The best imputation method is the **mode**, because it preserves existing category proportions.


```python
# Task 4: Remove duplicates

print("Number of duplicate rows:", medical_df.duplicated().sum())

# Droping duplicate rows
medical_df = medical_df.drop_duplicates()

medical_df.shape
```

## **Titanic.csv**

```python
# Problem 1: Subset Titanic DataFrame

print("Upload titanic.csv")
uploaded = files.upload()

titanic = pd.read_csv("titanic.csv")

# Subset only the required columns
subset_cols = ["Name", "Pclass", "Sex", "Age", "Fare", "Survived"]
df = titanic[subset_cols]

# Keeping only first-class passengers
df = df[df["Pclass"] == 1]

# Computing required stats
print("Fare Mean:", df["Fare"].mean())
print("Fare Median:", df["Fare"].median())
print("Fare Max:", df["Fare"].max())
print("Fare Min:", df["Fare"].min())
```

```python
# Problem 2: Null Age values

print("Null Age values:", df["Age"].isnull().sum())

# Drops rows where Age is null
df = df.dropna(subset=["Age"])

df.head()
```

```python
# Problem 3: One-hot encode Embarked

# Creating binary dummy variables
embarked_dummies = pd.get_dummies(titanic["Embarked"], prefix="Embarked")

# Adds to DataFrame
titanic_encoded = pd.concat([titanic, embarked_dummies], axis=1)

# Removing original column
titanic_encoded = titanic_encoded.drop(columns=["Embarked"])

titanic_encoded.head()
```

```python
# Problem 4: Survival rate by Gender

survival_by_gender = titanic.groupby("Sex")["Survived"].mean()
print(survival_by_gender)

plt.bar(survival_by_gender.index, survival_by_gender.values)
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Sex")
plt.show()
```

```python
# Problem 5: Survival rate by Embarked port

survival_by_port = titanic.groupby("Embarked")["Survived"].mean()
print(survival_by_port)

# Ploting survival rate by port
survival_by_port.plot(kind="bar")
plt.title("Survival Rate by Port of Embarkation")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Problem 6: Age quantiles + Pclass grouping

# Dividing Age into 5 quantile-based groups
titanic["AgeGroup"] = pd.qcut(titanic["Age"], q=5)

# Computing survival mean by Pclass & AgeGroup
table = titanic.groupby(["Pclass", "AgeGroup"], observed=True)["Survived"].mean()
print(table)

# Ploting the result
table.unstack().plot(kind="bar", figsize=(10,5))
plt.title("Survival by Pclass & AgeGroup")
plt.ylabel("Survival Rate")
plt.show()
```
